```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
results = json.load(open('rl-results-15000.json', 'r'))
```

```python
loss = results['loss']
idx, loss = zip(*loss)
plt.plot(idx, loss)
plt.show()
```

```python
r = results['rewards']
```

```python
tot_reward = [np.sum(r[str(x)]) for x in range(15000)]
plt.plot(tot_reward)
plt.show()
```

```python
num_pos = [np.sum(np.asarray(r[str(x)]) >=0) for x in range(15000)]
plt.plot(num_pos)
plt.show()
```

```python
samples = [1000, 2500, 5000, 7500, 10000, 15000]
for i in range(7):
    plt.plot(r[str((i+1)*2000)], label=(i+1)*2000)
plt.legend()
plt.show()
```
