```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
results = json.load(open('rl-results-15000.json', 'r'))
```

```python
loss = results['loss']
idx, loss = zip(*loss)
plt.plot(idx, loss)
plt.show()
```

```python
r = results['rewards']
```

```python
tot_reward = [np.sum(r[str(x)]) for x in range(15000)]
plt.plot(tot_reward)
plt.show()
```

```python
num_pos = [np.sum(np.asarray(r[str(x)]) >0) for x in range(15000)]
plt.plot(num_pos)
plt.show()
```
